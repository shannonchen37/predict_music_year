```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([[70,50],[35,90],[60,80]])
y = [35.8,50,70.4]
print(x[:, 0])
```

```python
# 实例化一个估计器
estimator = LinearRegression()
```

```python
estimator.fit(x,y)
```

```python
estimator.score(x,y)
```

```python
coef = estimator.coef_
print("系数：",coef)
w1 = coef[0]
w2 = coef[1]
```

```python
b = estimator.intercept_
print("截距：",b)
```

```python
pre = estimator.predict([[200,300]])
print("200,300的预测值是：",pre)
```

```python
# 画图
ax = plt.figure().add_subplot(111,projection = '3d')
ax.scatter(x[:, 0], x[:,1],y, c ="r",marker="o",linewidths=5)
x0 = x[:,0]
x1 = x[:,1]
x0 ,x1 = np.meshgrid(x0,x1)
z = b + w1 * x0 + w2 * x1
ax.plot_surface(x0, x1, z,color = "b")
ax.set_xlabel('area')
ax.set_ylabel('num_rooms')
ax.set_zlabel('price')
```
