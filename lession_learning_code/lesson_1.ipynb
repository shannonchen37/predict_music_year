```python
# https://blog.csdn.net/weixin_38828673/article/details/107574786
import numpy as np 
import random
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
```

```python
a=[random.randint(0,20) for x in range(1000)] #确定范围及个数
counter_res = Counter(a)
print("counter_res: ")
print(counter_res)
print("What is counter_res.items()?")
print(counter_res.items())
dic_res = {number: value for number, value in counter_res.items()}
print("dic_res: ")
print(dic_res)
print("dic_res.keys(): ")
print(dic_res.keys())
```

```python
# 取得key
x = [i for i in dic_res.keys()]
x.sort()
y = []
# 取得value
for i in x:
    y.append(dic_res.get(i))

print(x)
print(y)
```

```python
# 转换成dataFrame的格式
df = pd.DataFrame(y, x)
print(df)
```

```python
plt.xlabel = "number"
plt.ylabel = "count"
plt.plot(x, y)
```
