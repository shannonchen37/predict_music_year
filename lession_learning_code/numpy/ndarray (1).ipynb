```python
import matplotlib.pyplot as plt

#生成均匀分布的随机数
x1 = np.random.uniform(-1, 1, 1000)

#画布看分布情况
#1）创建画布
plt.figure(figsize=(5, 5), dpi=100) # dpi为解析度(每英寸的点数)

#绘制直方图
plt.hist(x=x1, bins=1000) 

#显示图像
plt.show();
```

```python
# 生成均匀分布的随机数
x1 = np.random.normal(loc=1.75, scale=1, size=100000000)

# 画图看分布状况
# 1）创建画布
plt.figure(figsize=(20, 10), dpi=100)

# 2）绘制直方图
plt.hist(x1, 1000)

# 3）显示图像
plt.show()
```
