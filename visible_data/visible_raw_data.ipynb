```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
import seaborn as sns
import numpy as np
```

```python
csv_path = "../datasets/data.csv"
```

```python
# 显示csv的前10行
# data = pd.read_csv(csv_path,nrows =5)
data = pd.read_csv(csv_path)
data.head(10)
```

```python
data.describe()
```

```python
# 读取Year的值,并统计出现次数
Year = []
for i in data.Year:
    Year.append(i)
n = collections.Counter(Year)
n
```

```python
# sort the dic base on year
dic = {number: value for number, value in n.items()}
dic_sort_tuple = sorted(dic.items())        # return a tuple
dic_sort = {k:v for k,v in dic_sort_tuple}  # transfor to a dic
dic_sort
```

```python
# sort the dic base on times
dic_sort_tuple = sorted(dic.items(),key=lambda x:x[1])  # return a tuple
dic_sort = {k:v for k,v in dic_sort_tuple}  # transfor to a dic
dic_sort
```

```python
year_range = 'Year >= 1922 & Year <= 2011'

f, ax = plt.subplots(figsize=(20, 9))
ax = sns.boxplot(x="Year", y='attr1', data=data.query(year_range))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(20, 9))
attr = []
for i in range(1,13):
    attr.append("attr" + str(i))
data_attr = data[attr]
ax=sns.boxplot(x="variable", y="value", data=pd.melt(data_attr))

plt.show()
```

```python
# 变量相关性图
corr=data_attr.corr()       #表示data中的两个变量之间的相关性
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)         #设置颜色
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,annot=True,annot_kws={'size':10},
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#group the songs by decade 
data['decade'] = data.Year.apply(lambda year : year-(year%10))
```

```python
decade = ["decade"]
for i in range(1,13):
    decade.append("attr" + str(i))

data_decade = data[decade]
columns = data_decade.groupby(['decade']).mean().columns
labels = ["{:02d}'s".format(l%100) for l in sorted(data_decade.decade.unique())]
fig, ax = plt.subplots(figsize=(10,5)) 
sns.heatmap(data_decade.groupby(['decade']).mean().iloc[:,0:20], yticklabels=labels)
plt.ylabel("Release Decade")
plt.xlabel("Features (Mean)")
plt.show()
```
