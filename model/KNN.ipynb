# 获取数据集；数据预处理；特征工程；机器学习；模型评估

```python
from sklearn.datasets import load_iris
```

```python
iris_datasets = load_iris()
# print(iris_datasets)
```

```python
print("数据集特征值: \n", iris_datasets.data)
print("数据集标签值: \n",iris_datasets['target'])
print("数据集特征值名称: \n",iris_datasets.feature_names)
print("数据集标签: \n",iris_datasets.target_names)
print("数据集描述: \n",iris_datasets.DESCR)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris_datasets_data = pd.DataFrame(data=iris_datasets.data, columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
iris_datasets_data["target"] = iris_datasets.target
print(iris_datasets_data)
```

```python
def iris_plot(data,col1,col2):
    sns.lmplot(x=col1,y=col2,data=data,hue="target")
    plt.title("Data of Iris")
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
iris_plot(iris_datasets_data,'sepal length (cm)', 'sepal width (cm)')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris_datasets.data,iris_datasets.target,test_size=0.3,train_size=0.7,random_state=20)
print("训练集特征值:\n",x_train)
print("测试集特征值:\n",x_test)
print("训练集标签值:\n",y_train)
print("测试集标签值:\n",y_test)
```

```python
print("size of training set:")
print(x_train.shape)
print("size of testing set:")
print(y_train.shape)
```

```python
# 特征预处理
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
# 归一化
# 存在的问题是如果存在一个数特别大，会导致归一化后其他数很小，因此往往采用标准化（MinMaxScaler）
transfer = MinMaxScaler(feature_range=(0,1))
minmax_data = transfer.fit_transform(iris_datasets_data[['target']])
print('归一化后为：\n', minmax_data)
```

```python
# 标准化
from sklearn.preprocessing import StandardScaler
transfer =  StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.fit_transform(x_test)
print(x_test)
```

```python
# mashine learning
from sklearn.neighbors import KNeighborsClassifier
estimator = KNeighborsClassifier()
estimator.fit(x_train,y_train)
```

```python
#  模型评估
y_pre = estimator.predict(x_test)
print("预测值：",y_pre)
print("预测值是否等于真实值",y_pre==y_test)

ret = estimator.score(x_test,y_test)
print("accuracy: ",ret)
```
